```python
import yt_dlp

def save_playlist_titles(playlist_url, output_file):
    ydl_opts = {'quiet': True}
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(playlist_url, download=False)
        titles = [video['title'] for video in info['entries']]
        
        with open(output_file, 'w', encoding='utf-8') as file:
            for idx, title in enumerate(titles, start=1):
                file.write(f"{idx}. {title}\n")
        
        print(f"Playlist titles saved to {output_file}")
```

```python
playlist_url = "https://www.youtube.com/playlist?list=PLmKXWmhsHrCapniOnL0FnEiXEnsoOfjdy"
output_file = "playlist_titles.txt"
save_playlist_titles(playlist_url, output_file)
```
